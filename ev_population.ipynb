```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\ADMIN\Downloads\Electric_Vehicle_Population_Data (1).csv")
df
```

```python
# Check exact column names and data types
print("Column names and data types:")
print(df.dtypes)
print("\nSample data:")
print(df.head(2))
```

```python
# Step 1: First, let's examine the current state of the data
print("Current column names and data types:")
print(df.dtypes)
print("\nMissing values per column:")
print(df.isnull().sum())

# Step 2: Rename columns first (if not already done)
column_rename = {
    'VIN (1-10)': 'vin_partial',
    'Postal Code': 'postal_code',
    'Model Year': 'model_year', 
    'Electric Vehicle Type': 'ev_type',
    'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'cafv_eligibility',
    'Electric Range': 'electric_range',
    'Base MSRP': 'base_msrp',
    'Legislative District': 'legislative_district',
    'DOL Vehicle ID': 'dol_vehicle_id',
    'Vehicle Location': 'vehicle_location',
    'Electric Utility': 'electric_utility',
    '2020 Census Tract': 'census_tract_2020'
}

# Only rename if columns exist with old names
existing_columns = {k: v for k, v in column_rename.items() if k in df.columns}
df = df.rename(columns=existing_columns)

print(f"\nRenamed {len(existing_columns)} columns")

# Step 3: Handle each numeric column separately with proper error handling

# For postal_code - keep as string, handle missing values
if 'postal_code' in df.columns:
    df['postal_code'] = df['postal_code'].fillna('Unknown')
    # Convert numeric postal codes to string, leave others as is
    df['postal_code'] = df['postal_code'].apply(lambda x: str(int(x)) if pd.notna(x) and str(x).replace('.','').isdigit() else str(x))
    print("Postal codes cleaned")

# For electric_range - handle missing values and convert to int
if 'electric_range' in df.columns:
    print(f"Electric Range - Missing: {df['electric_range'].isna().sum()}")
    print(f"Electric Range - Sample values: {df['electric_range'].unique()[:10]}")
    
    # Fill NaN with 0 and convert to int
    df['electric_range'] = pd.to_numeric(df['electric_range'], errors='coerce').fillna(0).astype(int)
    print("Electric range converted to int")

# For base_msRP - convert to int
if 'base_msrp' in df.columns:
    df['base_msrp'] = pd.to_numeric(df['base_msrp'], errors='coerce').fillna(0).astype(int)
    print("Base MSRP converted to int")

# For model_year - handle conversion carefully
if 'model_year' in df.columns:
    print(f"Model Year - Missing: {df['model_year'].isna().sum()}")
    df['model_year'] = pd.to_numeric(df['model_year'], errors='coerce')

  # If there are still missing values, fill with median
    if df['model_year'].isna().any():
        median_year = df['model_year'].median()
        df['model_year'] = df['model_year'].fillna(median_year)
    
    df['model_year'] = df['model_year'].astype(int)
    print("Model year converted to int")

# Step 4: Text standardization
text_columns = ['Make', 'Model', 'City', 'County', 'State']
for col in text_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
        if col in ['City', 'County']:
            df[col] = df[col].str.title()
        elif col in ['Make', 'Model']:
            df[col] = df[col].str.upper()
        print(f"{col} text standardized")

# Step 5: Simplify CAFV eligibility
if 'cafv_eligibility' in df.columns:
    df['cafv_eligibility'] = df['cafv_eligibility'].replace({
        'Not eligible due to low battery range': 'Not Eligible',
        'Clean Alternative Fuel Vehicle Eligible': 'Eligible', 
        'Eligibility unknown as battery range has not been researched': 'Unknown'
    })
    print("CAFV eligibility simplified")

# Step 6: Extract coordinates (optional - skip if causing issues)
try:
    import re
    
    def extract_coordinates(point_string):
        if pd.isna(point_string) or 'POINT' not in str(point_string):
            return None, None
        try:
            # Handle POINT(-122.90787 46.9461) format
            coords = re.findall(r'[-+]?\d*\.\d+|\d+', str(point_string))
            if len(coords) >= 2:
                return float(coords[0]), float(coords[1])
        except:
            pass
        return None, None

    if 'vehicle_location' in df.columns:
        coords = df['vehicle_location'].apply(extract_coordinates)
        df['longitude'] = coords.apply(lambda x: x[0] if x else None)
        df['latitude'] = coords.apply(lambda x: x[1] if x else None)
        print(f"Coordinates extracted: {df['longitude'].notna().sum()} valid points")
except Exception as e:
    print(f"Coordinate extraction skipped due to error: {e}")

# Step 7: Clean electric utility data
if 'electric_utility' in df.columns:
    df['electric_utility'] = df['electric_utility'].fillna('Unknown').astype(str)
    df['electric_utility'] = df['electric_utility'].str.replace(' - (WA)', '', regex=False)
    df['electric_utility'] = df['electric_utility'].str.replace('||', '|', regex=False)

    def categorize_utility(utility):
        utility = str(utility)
        if 'PUGET SOUND ENERGY' in utility:
            return 'Puget Sound Energy'
        elif 'CITY OF SEATTLE' in utility:
            return 'Seattle City Light'
        elif 'PACIFICORP' in utility:
            return 'PacifiCorp'
        elif 'PUD' in utility.upper():
            return 'Public Utility District'
        elif utility == 'Unknown' or utility == 'nan':
            return 'Unknown'
        else:
            return 'Other'
            
    df['utility_category'] = df['electric_utility'].apply(categorize_utility)
    print("Electric utility data cleaned")

# Step 8: Final validation
print("\n" + "="*60)
print("DATA CLEANING COMPLETED SUCCESSFULLY!")
print("="*60)

print(f"\nFinal dataset shape: {df.shape}")
print("\nData types after cleaning:")

print("\nMissing values after cleaning:")
print(df.isnull().sum())

print("\nSample of cleaned data:")
sample_cols = [col for col in ['vin_partial', 'model_year', 'Make', 'Model', 'electric_range', 'cafv_eligibility'] if col in df.columns]
print(df[sample_cols].head())

print(f"\nElectric Range statistics:")
print(f"Min: {df['electric_range'].min()}, Max: {df['electric_range'].max()}, Mean: {df['electric_range'].mean():.1f}")

print(f"\nModel Year range: {df['model_year'].min()} - {df['model_year'].max()}")


print(df.dtypes)



        




```

```python
# Vehicle make distribution
make_distribution = df['Make'].value_counts().head(10)
print("Top 10 EV Makes:")
print(make_distribution)

# EV type analysis
ev_type_analysis = df.groupby('ev_type').agg({
    'electric_range': 'mean',
    'model_year': 'mean',
    'vin_partial': 'count'
}).round(1)
print("\nEV Type Analysis:")
print(ev_type_analysis)

# Geographic distribution
county_distribution = df['County'].value_counts().head(5)
print("\nTop 5 Counties by EV Count:")
print(county_distribution)
```

```python
# EV adoption over time
yearly_trends = df.groupby('model_year').agg({
    'vin_partial': 'count',
    'electric_range': 'mean'
}).rename(columns={'vin_partial': 'vehicle_count'})

# Plot adoption trends
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
yearly_trends['vehicle_count'].plot(kind='bar', title='EV Adoption by Model Year')
plt.show()
```

```python
# Electric range analysis by vehicle type
range_analysis = df.groupby(['ev_type', 'Make']).agg({
    'electric_range': ['mean', 'max', 'count']
}).round(1)

# Filter for meaningful results (at least 10 vehicles)
range_analysis = range_analysis[range_analysis[('electric_range', 'count')] >= 10]
print("Range Analysis by Make and EV Type:")
print(range_analysis.sort_values(('electric_range', 'mean'), ascending=False).head(10))
```

```python
# Regional analysis
regional_analysis = df.groupby('County').agg({
    'vin_partial': 'count',
    'electric_range': 'mean',
    'model_year': 'mean'
}).round(1).sort_values('vin_partial', ascending=False)

print("Regional EV Distribution:")
print(regional_analysis.head(10))

# Utility provider analysis
utility_analysis = df['utility_category'].value_counts()
print("\nEV Distribution by Utility Provider:")
print(utility_analysis)
```

```python
# Eligibility patterns
cafv_analysis = df.groupby(['ev_type', 'cafv_eligibility']).size().unstack(fill_value=0)
print("CAFV Eligibility by EV Type:")
print(cafv_analysis)

# Eligibility vs electric range
eligibility_range = df.groupby('cafv_eligibility')['electric_range'].describe()
print("\nElectric Range by CAFV Eligibility:")
print(eligibility_range)
```

```python
# Validate the cleaned data
print("Data Validation Checks:")
print(f"Total records: {len(df):,}")
print(f"Unique VINs: {df['vin_partial'].nunique():,}")
print(f"Duplicate check: {len(df) - df['vin_partial'].nunique()} potential duplicates")
print(f"Valid coordinates: {df['latitude'].notna().sum():,}")
print(f"Model year validity: {len(df[df['model_year'] > 2024])} future model years")

# Check for data anomalies
anomalies = df[df['electric_range'] == 0]
print(f"\nVehicles with 0 electric range: {len(anomalies):,}")
print("Breakdown by EV type:")
print(anomalies['ev_type'].value_counts())
```

```python
import pandas as pd

print("=== EXECUTING ALL FIVE ANALYSES AND EXPORTING TO EXCEL ===")

# Analysis 1: Basic Descriptive Analysis
print("1. Running Basic Descriptive Analysis...")
make_distribution = df['Make'].value_counts().head(15).reset_index()
make_distribution.columns = ['Make', 'Vehicle_Count']

ev_type_analysis = df.groupby('ev_type').agg({
    'electric_range': ['mean', 'median', 'min', 'max'],
    'model_year': ['mean', 'min', 'max'],
    'vin_partial': 'count'
}).round(1)
ev_type_analysis.columns = ['avg_range', 'median_range', 'min_range', 'max_range', 
                           'avg_year', 'min_year', 'max_year', 'vehicle_count']
ev_type_analysis = ev_type_analysis.reset_index()

county_distribution = df['County'].value_counts().head(10).reset_index()
county_distribution.columns = ['County', 'Vehicle_Count']

# Export Analysis 1 to Excel
with pd.ExcelWriter('Analysis1_Descriptive_Statistics.xlsx') as writer:
    make_distribution.to_excel(writer, sheet_name='Top_Makes', index=False)
    ev_type_analysis.to_excel(writer, sheet_name='EV_Type_Analysis', index=False)
    county_distribution.to_excel(writer, sheet_name='Top_Counties', index=False)

print(" Analysis 1 exported: Analysis1_Descriptive_Statistics.xlsx")

# Analysis 2: Temporal Trends
print("2. Running Temporal Trends Analysis...")
yearly_trends = df.groupby('model_year').agg({
    'vin_partial': 'count',
    'electric_range': 'mean',
    'base_msrp': 'mean'
}).round(1).reset_index()
yearly_trends.columns = ['model_year', 'vehicle_count', 'avg_electric_range', 'avg_msrp']

# Calculate year-over-year growth
yearly_trends['yoy_growth'] = yearly_trends['vehicle_count'].pct_change() * 100

# Export Analysis 2 to Excel
yearly_trends.to_excel('Analysis2_Temporal_Trends.xlsx', index=False)
print(" Analysis 2 exported: Analysis2_Temporal_Trends.xlsx")

# Analysis 3: Range Analysis by Vehicle Type
print("3. Running Range Analysis by Vehicle Type...")
range_analysis = df.groupby(['ev_type', 'Make']).agg({
    'electric_range': ['mean', 'max', 'min', 'count']
}).round(1)
range_analysis.columns = ['avg_range', 'max_range', 'min_range', 'vehicle_count']
range_analysis = range_analysis.reset_index()

# Filter for meaningful results (at least 5 vehicles)
range_analysis_filtered = range_analysis[range_analysis['vehicle_count'] >= 5]

# Top performers by EV type
top_range_bev = range_analysis_filtered[range_analysis_filtered['ev_type'] == 'Battery Electric Vehicle (BEV)'].nlargest(10, 'avg_range')
top_range_phev = range_analysis_filtered[range_analysis_filtered['ev_type'] == 'Plug-in Hybrid Electric Vehicle (PHEV)'].nlargest(10, 'avg_range')

# Export Analysis 3 to Excel
with pd.ExcelWriter('Analysis3_Range_Analysis.xlsx') as writer:
    range_analysis_filtered.to_excel(writer, sheet_name='All_Makes_Range', index=False)
    top_range_bev.to_excel(writer, sheet_name='Top_BEV_Range', index=False)
    top_range_phev.to_excel(writer, sheet_name='Top_PHEV_Range', index=False)

print(" Analysis 3 exported: Analysis3_Range_Analysis.xlsx")

# Analysis 4: Geographic Insights
print("4. Running Geographic Insights Analysis...")
regional_analysis = df.groupby('County').agg({
    'vin_partial': 'count',
    'electric_range': 'mean',
    'model_year': 'mean',
    'base_msrp': 'mean'
}).round(1).sort_values('vin_partial', ascending=False).reset_index()
regional_analysis.columns = ['county', 'vehicle_count', 'avg_range', 'avg_model_year', 'avg_msrp']

# Utility provider analysis
utility_analysis = df.groupby('utility_category').agg({
    'vin_partial': 'count',
    'electric_range': 'mean',
    'model_year': 'mean'
}).round(1).sort_values('vin_partial', ascending=False).reset_index()
utility_analysis.columns = ['utility', 'vehicle_count', 'avg_range', 'avg_model_year']

# City-level analysis (top 15 cities)
city_analysis = df.groupby('City').agg({
    'vin_partial': 'count',
    'electric_range': 'mean'
}).round(1).nlargest(15, 'vin_partial').reset_index()
city_analysis.columns = ['city', 'vehicle_count', 'avg_range']

# Export Analysis 4 to Excel
with pd.ExcelWriter('Analysis4_Geographic_Insights.xlsx') as writer:
    regional_analysis.to_excel(writer, sheet_name='County_Analysis', index=False)
    utility_analysis.to_excel(writer, sheet_name='Utility_Analysis', index=False)
    city_analysis.to_excel(writer, sheet_name='Top_Cities', index=False)

print(" Analysis 4 exported: Analysis4_Geographic_Insights.xlsx")

# Analysis 5: CAFV Eligibility Analysis
print("5. Running CAFV Eligibility Analysis...")
cafv_analysis = df.groupby(['ev_type', 'cafv_eligibility']).size().reset_index()
cafv_analysis.columns = ['ev_type', 'cafv_eligibility', 'vehicle_count']
cafv_pivot = cafv_analysis.pivot_table(
    index='ev_type', 
    columns='cafv_eligibility', 
    values='vehicle_count', 
    fill_value=0
).reset_index()

# Eligibility vs electric range
eligibility_range = df.groupby('cafv_eligibility').agg({
    'electric_range': ['mean', 'median', 'min', 'max', 'count']
}).round(1)
eligibility_range.columns = ['avg_range', 'median_range', 'min_range', 'max_range', 'vehicle_count']
eligibility_range = eligibility_range.reset_index()

# Eligibility by model year
eligibility_by_year = df.groupby(['model_year', 'cafv_eligibility']).size().reset_index()
eligibility_by_year.columns = ['model_year', 'cafv_eligibility', 'vehicle_count']

# Export Analysis 5 to Excel
with pd.ExcelWriter('Analysis5_CAFV_Eligibility.xlsx') as writer:
    cafv_pivot.to_excel(writer, sheet_name='Eligibility_by_EV_Type', index=False)
    eligibility_range.to_excel(writer, sheet_name='Range_by_Eligibility', index=False)
    eligibility_by_year.to_excel(writer, sheet_name='Eligibility_Trends', index=False)

print(" Analysis 5 exported: Analysis5_CAFV_Eligibility.xlsx")

# Bonus: Create a comprehensive summary dashboard
print("6. Creating Comprehensive Summary Dashboard...")
summary_stats = {
    'Metric': [
        'Total Vehicles', 'Unique Makes', 'Unique Models', 'Counties Covered',
        'Average Electric Range', 'Max Electric Range', 'Average Model Year',
        'BEV Percentage', 'PHEV Percentage', 'Eligible Percentage'
    ],
    'Value': [
        len(df),
        df['Make'].nunique(),
        df['Model'].nunique(), 
        df['County'].nunique(),
        round(df['electric_range'].mean(), 1),
        df['electric_range'].max(),
        round(df['model_year'].mean(), 1),
        f"{(df['ev_type'] == 'Battery Electric Vehicle (BEV)').mean() * 100:.1f}%",
        f"{(df['ev_type'] == 'Plug-in Hybrid Electric Vehicle (PHEV)').mean() * 100:.1f}%",
        f"{(df['cafv_eligibility'] == 'Eligible').mean() * 100:.1f}%"
    ]
}

summary_df = pd.DataFrame(summary_stats)

# Export Summary Dashboard
with pd.ExcelWriter('EV_Analysis_Summary_Dashboard.xlsx') as writer:
    summary_df.to_excel(writer, sheet_name='Key_Metrics', index=False)
    make_distribution.to_excel(writer, sheet_name='Top_Makes', index=False)
    yearly_trends.to_excel(writer, sheet_name='Yearly_Trends', index=False)
    regional_analysis.head(10).to_excel(writer, sheet_name='Top_Counties', index=False)

print(" Summary Dashboard exported: EV_Analysis_Summary_Dashboard.xlsx")

print("\n" + "="*60)
print(" ALL ANALYSES COMPLETED AND EXPORTED!")
print("="*60)
print("\n Generated Excel Files:")
print("1. Analysis1_Descriptive_Statistics.xlsx")
print("2. Analysis2_Temporal_Trends.xlsx") 
print("3. Analysis3_Range_Analysis.xlsx")
print("4. Analysis4_Geographic_Insights.xlsx")
print("5. Analysis5_CAFV_Eligibility.xlsx")
print("6. EV_Analysis_Summary_Dashboard.xlsx (Bonus)")
print(f"\n Total records analyzed: {len(df):,} vehicles")
```
